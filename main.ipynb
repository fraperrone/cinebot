1. Sistema de reservas para un cine

Un cine quiere modernizar su atención al cliente creando un pequeño sistema en Python
que permita reservar entradas, consultar funciones y dejar opiniones sobre las películas.
Crearemos un prototipo simple de asistente virtual para un cine.


```python
print(""" 
Hola, soy CineBot, tu asistente virtual del cine MovieTime"
Indicame cómo te puedo ayudar:
1. Reservar entradas
2. Consultar cartelera
3. Dejar una reseña
"Ingrese la opción:
""")
opcion = input()
```

Opcion 1: Reserva de entradas

```python
if opcion == "1":
    print("""
    Tenemos las siguientes películas disponibles:
    1. Argentina, 1985
    2. El secreto de sus ojos
    3. Relatos salvajes
    ¿Qué película querés ver?
    """)
    pelicula = input()
    precio_tipo_entrada = {
        "2D": 3500,
        "3D": 4500,
        "VIP": 6000
    }
    print("""
    ¿Qué tipo de entrada desea?
    1. 2D
    2. 3D
    3. VIP    
    """)
    tipo_entrada = input()
    print("Cuantas quieres comprar?")
    cantidad = int(input())
    personas = []
    for i in range(cantidad):
        print(f"Ingrese el nombre del espectador {i+1}:")
        nombre = input()
        personas.append(nombre)

    precio_total = precio_tipo_entrada[tipo_entrada] * cantidad
    print(f"El total a pagar es: ${precio_total}")
    print("Con que nombre desea hacer la reserva?")
    nombre_reserva = input()
    print(f"Gracias por utilizar MovieTime, {nombre_reserva.upper()}. Tu reserva fue realizada con éxito. Te esperamos en el cine!")
```

Opción 2: Consultar cartelera

```python
if opcion == "2":
    print("""
    Cartelera:
    1. Argentina, 1985
    2. El secreto de sus ojos
    3. Relatos salvajes
    """)
    print("Gracias por consultar la cartelera de MovieTime. Te esperamos en el cine!")
```

Opción 3: Dejar una reseña

```python
if opcion == "3":
    print("""
    Qué película querés reseñar?
    1. Argentina, 1985
    2. El secreto de sus ojos
    3. Relatos salvajes
 """)
    pelicula_reseñar = input()
    print("Escribe tu reseña:")
    reseña = input()
    print("del 1 al 5, ¿qué puntuación le das a la película?")
    puntuacion = int(input())
    if puntuacion == 5:
        print("Nos alegra que hayas disfrutado la película al máximo! Gracias por tu reseña.")
    elif puntuacion == 4 or puntuacion == 3:
        print("Gracias por tu comentario. Seguiremos mejorando.")
    elif puntuacion == 2 or puntuacion == 1:
        print("Lamentamos que tu experiencia no haya sido la mejor")
```
